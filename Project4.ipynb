```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import nltk
nltk.download("vader_lexicon")
import re
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sent_analyzer = SentimentIntensityAnalyzer()
from sklearn.metrics import accuracy_score, classification_report
```

```python
data_df = pd.read_csv('Resources/0819_UkraineCombinedTweetsDeduped.csv')
data_df.info()
```

```python
#Popular tweet by retweet
bins = [-1,0,300,700, 1000]
labels = ['No Popularity','Low Popularity','Medium Popularity','High Popularity']
data_df['Popularity'] = pd.cut(x=data_df['retweetcount'],bins=bins,labels=labels)
data_df
```

```python
polarity_scores = []
for text in data_df["text"]:
    score = sent_analyzer.polarity_scores(text)["compound"]
    polarity_scores.append(score)
data_df["polarity_score"] = polarity_scores

# Create a new column with the sentiment categories
sentiment_categories = []
for score in data_df["polarity_score"]:
    if score >= 0.05:
        category = "positive"
    elif score <= -0.05:
        category = "negative"
    else:
        category = "neutral"
    sentiment_categories.append(category)
data_df["sentiment_category"] = sentiment_categories
```

```python
sentiments = []
for i in range(len(data_df)):
    text = data_df.iloc[i]["text"]
    scores = sent_analyzer.polarity_scores(text)
    max_score = max(scores, key=scores.get)
    if max_score == "pos":
        sentiments.append("positive")
    elif max_score == "neg":
        sentiments.append("negative")
    else:
        sentiments.append("neutral")
data_df["dominant_sentiment"] = sentiments
```

```python
vect = CountVectorizer(ngram_range=(1,2)).fit(data_df['text'])
```

```python
feature_names = vect.get_feature_names()
print("Number of features: {}\n".format(len(feature_names)))
print("First 20 features:\n {}".format(feature_names[:20]))
```

```python
X = data_df['text']
Y = data_df['sentiment_category']
X = vect.transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print("Size of x_train:", (x_train.shape))
print("Size of y_train:", (y_train.shape))
print("Size of x_test:", (x_test.shape))
print("Size of y_test:", (y_test.shape))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(logreg_pred, y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
print(confusion_matrix(y_test, logreg_pred))
print("\n")
print(classification_report(y_test, logreg_pred))
```

```python
style.use('classic')
cm = confusion_matrix(y_test, logreg_pred, labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=logreg.classes_)
disp.plot()
```
