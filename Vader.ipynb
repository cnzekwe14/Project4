```python
#import dependencies
import pandas as pd
import nltk
from sklearn.metrics import accuracy_score, classification_report
```

```python
#connect vader
nltk.download("vader_lexicon")
```

```python
#import SentimentIntensityAnalyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sent_analyzer = SentimentIntensityAnalyzer()
```

```python
#read in data
data = pd.read_csv('0819_UkraineCombinedTweetsDeduped.csv')
```

```python
data.head()
```

```python
#drop columns
data_df = data.drop(['original_tweet_userid', 'acctdesc', 'location', 'coordinates', 'original_tweet_username', 'in_reply_to_status_id', 'in_reply_to_user_id','in_reply_to_screen_name', 'is_quote_status', 'quoted_status_id', 'quoted_status_userid',
       'quoted_status_username'], axis=1)
```

```python
data_df
```

```python
#add new column called Popularity
bin_label = ['No Popularity','Low Popularity','Medium Populatrity','High Popularity']
bin_cut = [-1,0,150,300,1000]
data_df['Popularity'] = pd.cut(data_df['retweetcount'],bins=bin_cut,labels=bin_label)
data_df.head()
```

```python
polarity_scores = []
for text in data_df["text"]:
    score = sent_analyzer.polarity_scores(text)["compound"]
    polarity_scores.append(score)
data_df["polarity_score"] = polarity_scores
```

```python
def format_output(output_1):
    polarity = "neutral"
    
    if(output_1['compound']>=.05):
        polarity = "positive"
    elif(output_1['compound']<=-.05):
        polarity = "negative"
    return polarity
def predict_1(text):
    output_dict = sent_analyzer.polarity_scores(text)
    return format_output(output_dict)

data_df['sentiment_category'] = data_df['text'].apply(predict_1)


```

```python
sentiments = []
for i in range(len(data_df)):
    text = data_df.iloc[i]["text"]
    scores = sent_analyzer.polarity_scores(text)
    max_score = max(scores, key=scores.get)
    if max_score == "pos":
        sentiments.append("positive")
    elif max_score == "neg":
        sentiments.append("negative")
    else:
        sentiments.append("neutral")
data_df["dominant_sentiment"] = sentiments
```

```python
data_df.rename(columns={'polarity_score':'compound'})
```

```python
#using compound score
data_df[['sentiment_category','Popularity']].value_counts()
```

```python
#max value of polarity (positive, negative and neutral)
data_df[['dominant_sentiment','Popularity']].value_counts()
```

```python
accuracy = accuracy_score(data_df['dominant_sentiment'], data_df['sentiment_category'])

print("Accuracy: {}\n".format(accuracy))

# Show the classification report
print(classification_report(data_df['dominant_sentiment'], data_df['sentiment_category']))
```
